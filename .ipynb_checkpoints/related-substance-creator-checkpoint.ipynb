```python
import os
import re
from tqdm import tqdm
from datetime import datetime

import pandas as pd
import numpy as np
import camelot
import tabula
```

```python
def shift_row_to_top(df, index_to_shift):
    idx = df.index.tolist()
    idx.remove(index_to_shift)
    df = df.reindex([index_to_shift] + idx)
    return df
```

```python
def check_cols(columns, pdf_type):
    if(pdf_type == 'chrom'):
        if('Name' in columns and 'Area' in columns):
            return True
        else:
            return False
    elif(pdf_type == 'area'):
        if('Tailing Factor' in columns and 'Area' in columns):
            return True
        else:
            return False
```

```python
compound = 'ketorolacTromethamine'
# Bumetanide
# Acyclovir
# Famotidine
# ketorolacTromethamine
# KetorolacTromethamine
# LabetalolHCl
chrom_input = os.path.join(os.getcwd(), "data", "input", "{}-chromatogram.pdf".format(compound))
tables = camelot.read_pdf(chrom_input, pages= 'all',flavor='stream')
```

```python
df_peak_table =''
peak_tables = []
for table in tables:
    df_table = table.df
    search = df_table.where(df_table=='Peak#').dropna(how='all').dropna(axis=1)
    inx = list(search.index)
    if(inx):
        inx= inx[0]
        new_header = df_table.iloc[inx]
        new_start_inx = inx+1
        df_table = df_table[new_start_inx:] 
        df_table.columns = new_header
        df_table = df_table[['Peak#','Name','Ret. Time','Area','Area%','RRT']]
        peak_tables.append(df_table)
    else:
        continue
df_peak_table = pd.concat(peak_tables, ignore_index=True)            
inx_to_shift = df_peak_table[df_peak_table["Name"].str.contains('ketorolacTromethamine', flags = re.IGNORECASE)].index[0]
df_peak_table = shift_row_to_top(df_peak_table, inx_to_shift)
```

```python
compound = 'Acyclovir'
# Bumetanide
# Acyclovir
# Famotidine
# ketorolacTromethamine
# KetorolacTromethamine
# LabetalolHCl
area_input = os.path.join(os.getcwd(), "data", "input", "{}-areas.pdf".format(compound))
tables = camelot.read_pdf(area_input, pages= 'all')
```

```python
df_area_table =''
area_tables = []
for table in tables:
        df_table = table.df
        new_header = df_table.iloc[0]
        df_table = df_table[1:] 
        df_table.columns = new_header
        cols = list(df_table.columns)
        if(check_cols(cols, "area")):
            area_tables.append(df_table)
        else:
            continue
df_area_table = pd.concat(area_tables, ignore_index=True)
average_area = int(df_area_table["Area"][df_area_table["Title"] == "Average"].values.tolist()[0])
```
